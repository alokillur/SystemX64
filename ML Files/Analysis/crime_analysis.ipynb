```python
def test(crimetype, state, district, year, gender, season, age):
    state_data = cData[cData['STATE/UT'] == state]
    if state_data.empty:
        return "State not found in the dataset."

    top_districts = []
    for district in state_data['DISTRICT'].unique():
        district_data = state_data[state_data['DISTRICT'] == district]
        crime_count = district_data[crimetype].sum()
        top_districts.append((district, crime_count))

    top_districts.sort(key=lambda x: x[1], reverse=True)
    top_5_districts = top_districts[:5]

    return top_5_districts

result = test("RAPE", "ANDHRA PRADESH","ANANTAPUR", 2011, "Male", "STRIKE", 52)
print(result)
```
